   **DOMAIN ANALYSIS OF KEY  FEATURES USED AS OBTAINED FROM PROVIDED LOAN DICTIONARY.**
   
1.`Term`-The length of the loan expressed in months.

2.`BorrowerAPR`-The Borrower's Annual Percentage Rate (APR) for the loan.

3.`ProsperScore`-A custom risk score built using historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score.  Applicable for loans originated after July 2009.

4.`LenderYield`-The Lender yield on the loan. Lender yield is equal to the interest rate on the loan less the servicing fee.

5.`LoanOriginalAmount`-The origination amount of the loan.

6.`LoanStatus`-The current status of the loan: Cancelled,  Chargedoff, Completed, Current, Defaulted, FinalPaymentInProgress, PastDue. The PastDue status will be accompanied by a delinquency bucket.

7.`ProsperRating_Alpha`-The Prosper Rating assigned at the time the listing was created between AA - HR.  Applicable for loans originated after July 2009.

8.`ListingCategory_numeric`-The category of the listing that the borrower selected when posting their listing: 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans

9.`Occupation`-The Occupation selected by the Borrower at the time they created the listing.

10.`IncomeRange`-The income range of the borrower at the time the listing was created.

**We are going to convert ProsperingRating_Alpha, IncomeRange to Ordinal categorical data.**

```python
#Converting ProsperRating_Alpha and IncomeRange to ordinal categorical datatype. 
ordinal_variable_dictionary = {'ProsperRating_Alpha': ['N/A','HR','E','D','C', 'B', 'A', 'AA'],\
                    'IncomeRange':['$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+']}

for var in ordinal_variable_dictionary:
    ordered_variable = pd.api.types.CategoricalDtype(ordered = True,categories = ordinal_variable_dictionary[var])
    new_loans_df[var] = new_loans_df[var].astype(ordered_variable)
```

**We are use the map method to replace the Listingcategoric_numeric numbers with their respective names.**

```python
#Replacing listingcategory_numeric to their respective names using the map method.
category_number_name = {0:  'Not Available', 1 : 'Debt Consolidation', 2 : 'Home Improvement', 3 : 'Business', 4 : 'Personal Loan',\
                5 : 'Student Use', 6 : 'Auto', 7 : 'Other', 8 : 'Baby&Adoption', 9 : 'Boat', 10 : 'Cosmetic Procedure',\
                11 : 'Engagement Ring', 12 : 'Green Loans', 13 : 'Household Expenses', 14 : 'Large Purchases',\
                15 : 'Medical/Dental', 16 : 'Motorcycle', 17 : 'RV', 18 : 'Taxes', 19 : 'Vacation', 20 : 'Wedding Loans'}

new_loans_df['ListingCategory_numeric'] = new_loans_df['ListingCategory_numeric'].map(category_number_name)
```

**We then adjust the categorical data types to the correct format before analysis.**

```python
new_loans_df.Occupation = new_loans_df.Occupation.astype('category')
new_loans_df.LoanStatus = new_loans_df.LoanStatus.astype('category')
new_loans_df.ListingCategory_numeric = new_loans_df.ListingCategory_numeric.astype('category')
```

__Listing Category distribution investigation__

```python
# extracting unique values from ListingCategory_numeric feature. and odering them in a decreasing order. 
Category_counts = Master_df['ListingCategory_numeric'].value_counts()
Category_order=Category_counts.index

# Returns the sum of all not-null values in `ListingCategory_numeric feature
Category_df = Master_df['ListingCategory_numeric'].value_counts().sum()

# extracting the [0]th index which is the highest in the ListingCategory_numeric feature.
maximum_Category_counts =Category_counts[0]

# getting the maximum proportion of the bars to be plotted.
maximum_prop = maximum_Category_counts / Category_df

#creating category proprtions usin arange method from numpy library
Category_proportions = np.arange(0, maximum_prop, 0.04)

# Plot the ListingCategory on a Horizontal bar chart
plt.figure(figsize = (20, 10))
base_color = sns.color_palette()[2]
sns.countplot(data=Master_df, y='ListingCategory_numeric', color=base_color, order=Category_order);
plt.title('ListingCategory distribution');

# we use the for loop to print the proportion texts to our plotted bars.This will be printed in descending order.
for i in range (Category_counts.shape[0]):
    count =Category_counts[i]
    # Converting the count values into  percentagesthen into string
    percentage_str = '{:0.1f}'.format(100*count/Category_df)
    # Printing the string values to their respective bars. 
    plt.text(count+1, i, percentage_str, va='center',color='r')
```
